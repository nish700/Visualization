```python
#check decimal to percentage
deforest_data_1 = deforest_data.sort_values("Primary / Total tree cover (2018)").reset_index()
deforest_data_1['% tree cover (2018)'] = deforest_data['% tree cover (2018)'] * 100
deforest_data_1['Primary / Total tree cover (2018)'] = deforest_data_1['Primary / Total tree cover (2018)'] * 100
deforest_data_1['% loss\n(2001-2018)'] = deforest_data_1['% loss\n(2001-2018)'] * 100
```

```python
#plotting the interactive scatter plot
fig = px.scatter(
    deforest_data_1,
    x='Country',
    y = '% tree cover (2018)',
    size = 'Primary / Total tree cover (2018)',
    hover_name = 'Country',
    color='Country',
#     log_x = True,
    size_max = 40
)

app.layout = html.Div([dcc.Graph(id='Country wise tree cover', figure=fig)])

if __name__ == "__main__":
    app.run_server(debug=False)
```
