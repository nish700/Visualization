# Word Cloud for English Corpus

### Importing Libraries

```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

from wordcloud import WordCloud, STOPWORDS
from string import punctuation

from PIL import Image
import re
```

### uploading the English text

```python
#reading the contents of the text file
text_corpus=open('/content/iiit_overview_eng.txt', 'r').read()
```

```python
#checking the text corpus
text_corpus
```

### Stopwords

```python
#initializing the stopword
stopwords = set(STOPWORDS)
```

```python
#defining the custom stopwords
custom_stopwords = list(punctuation) + list(stopwords)
```

### Drawing the wordcloud

```python
# instantiate a word cloud object
iiit_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=custom_stopwords
)

# generate the word cloud
iiit_wc.generate(text_corpus)

# display the word cloud
plt.imshow(iiit_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#define the size of the figure
fig = plt.figure()
fig.set_figwidth(16) 
fig.set_figheight(18)

# display the cloud
plt.imshow(iiit_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Word Cloud in shape of Mask image

```python
mask = np.array(Image.open(r'/content/mask_image.jpg'))
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(13)

plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
mask_wc = WordCloud(
    stopwords = custom_stopwords,
    mask=mask,
    background_color='white',
    max_words = 2000,
    max_font_size=500,
    random_state=42
)

mask_wc.generate(text_corpus)
plt.figure(figsize=(20,10))
plt.imshow(mask_wc, interpolation='None')
plt.axis('off')
plt.show()
```

# Word Cloud for Hindi Corpus

### Importing hindi text

```python
#reading the contents of the hindi text file
hindi_corpus=open('/content/hindi_text.txt', 'r').read()
```

```python
hindi_corpus
```

### defining the stopwords

```python
stopword=open('/content/stopwords_hindi.txt','r')
stop_words=[]
# pre-process stopword
for i in stopword:
  i = re.sub('[\n]', '', i)
  stop_words.append(i)
  
stopwords_hindi = set(stop_words)
```

```python
custom_stopwords_hindi = list(punctuation) + list(stopwords_hindi)
```

### Importing map of India

```python
india_mask = np.array(Image.open(r'/content/india.jpg'))
#display the logo
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(8)

plt.imshow(india_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Word Cloud in the Shape of India Map

```python
mask_india = WordCloud(
    font_path='/content/Lohit-Devanagari.ttf',
    stopwords = custom_stopwords_hindi,
    mask=india_mask,
    background_color='white',
    max_words = 2000,
    max_font_size=700,
    random_state=42
)

mask_india.generate(hindi_corpus)
plt.figure(figsize=(20,15))
plt.imshow(mask_india, interpolation='None')
plt.axis('off')
plt.show()
```
